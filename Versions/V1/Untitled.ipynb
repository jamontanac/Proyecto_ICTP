```python
import numpy as np
import matplotlib.pylab as plt
import os
```

```python
names=[i for i in os.listdir(".") if i.endswith(".txt")]
b=np.array([])
aux=0
for i in range(len(names)):
    print("matrix_"+str(i)+".txt")
    if aux==0:
        b=np.genfromtxt("matrix_"+str(i)+".txt")
    else:
        a=np.genfromtxt("matrix_"+str(i)+".txt")
        b=np.vstack((b,a))
    aux+=1
    #b=np.vstack((b,a))
```
